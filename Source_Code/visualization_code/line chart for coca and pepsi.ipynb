```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set()
```

```python
trend = pd.read_csv("/Users/zym/Documents/bigdata/project/trend.csv")
trend.head()
```

```python
sns.set(font_scale=2)
# Initialize a grid of plots with an Axes for each walk
grid = sns.FacetGrid(trend, col="brand", hue="brand", col_wrap=4, size=6)

# Draw a horizontal line to show the starting point
grid.map(plt.axhline, y=0, ls=":", c=".5")

# Draw a line plot to show the trajectory of each random walk
grid.map(plt.plot, "month", "ratio=positive:negative", marker="o", ms=5 )


# Adjust the tick positions and labels
grid.set(xticks=np.arange(12), yticks=[0,1,2,3],
         xlim=(0, 12), ylim=(1, 3))

# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=1)

grid.savefig('/Users/zym/Documents/bigdata/project/trend.png', dpi=200, bbox_inches='tight')
```
